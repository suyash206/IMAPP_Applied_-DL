###Answer

1. The Row Vector (2D)
    A true row vector is a 2D array with shape (1, n).

    Creation: vec.reshape(1, -1) or vec[np.newaxis, :].

    Shape: (1, 3) — one row, three columns.

2. The Column Vector (2D)
    A column vector is a 2D array with shape (n, 1).

    Creation: vec.reshape(-1, 1) or vec[:, np.newaxis].

    Shape: (3, 1) — three rows, one column.

### Row vs. Column Vectors in NumPy

| Feature | Row Vector | Column Vector |
| :--- | :--- | :--- |
| **Shape** | `(1, N)` | `(N, 1)` |
| **Creation** | `vec.reshape(1, -1)` | `vec.reshape(-1, 1)` |
| **Indexing** | `arr[0, i]` | `arr[i, 0]` |
| **Math Role** | Left operand ($v \cdot A$) | Right operand ($A \cdot v$) |
| **Broadcasting** | Stretches vertically | Stretches horizontally |
| **Transpose** | Becomes Column Vector | Becomes Row Vector |





```python
import numpy as np

vec = np.array([1, 2, 3, 4, 5, 6])

# --- EXERCISE 1: Row or Column ---

# 1. Reshaping
a = vec.reshape(-1, 1)  # Works: Creates (6, 1) column vector
b = vec.reshape(1, -1)  # Works: Creates (1, 6) row vector
c = vec.reshape(2, 3)  # Works: Creates (2, 3) matrix
d = vec.reshape(-1, 3) # Works: Creates (2, 3) matrix (inferred rows)
# d = vec.reshape(-1, -1) # ERROR: Can only specify one unknown dimension

# 2. New Axis / Indexing
e = vec[..., np.newaxis] # Works: (6, 1) column vector
f = c[:, np.newaxis]    # Works: (2, 1, 3) tensor
g = c[np.newaxis, :, :] # Works: (1, 2, 3) tensor
h = c[np.newaxis, ...]  # Works: (1, 2, 3) tensor

# 3. Products and Multiplications
i = np.dot(vec, vec)    # Works: Scalar inner product (91)
j = vec @ vec           # Works: Scalar inner product (91)

# 4. Transpose
k = vec.T               # Works but UNCHANGED: 1D array remains (6,)
l = a.T                 # Works: Transposes (6, 1) to (1, 6)

# 5. Matrix Math Rules
# m = np.dot(a, vec)    # ERROR: (6,1) dot (6,) - inner dims mismatch
# n = c @ d             # ERROR: (2,3) @ (2,3) - inner dims mismatch
o = c.T @ d             # Works: (3,2) @ (2,3) -> (3,3) matrix
p = c @ d.T             # Works: (2,3) @ (3,2) -> (2,2) matrix
q = c @ d.T @ c         # Works: (2,2) @ (2,3) -> (2,3) matrix
r = np.dot(c, d.T)      # Works: (2,3) dot (3,2) -> (2,2) matrix
s = np.matmul(c, d.T)   # Works: Same as @, (2,2) result

# 6. Scalar multiplication
t = np.dot(c, 4)        # Works: Each element * 4
u = np.dot(4, c)        # Works: Each element * 4
v = 4 * c               # Works: Element-wise scalar multiplication

print(f"i: {i}, q shape: {q.shape}, v first element: {v[0,0]}")
```

### Definition: Exercise 1 - Matrix Operations
- `reshape(-1, 1)`: Creates a column vector by automatically calculating the number of rows.
- `reshape(1, -1)`: Creates a row vector.
- `@` or `np.dot`: Performs matrix multiplication. For 1D arrays, it computes the inner product.
- `.T`: Transposes the array. Note that 1D arrays like `(6,)` remain unchanged; you must convert them to 2D matrices first.

### Detailed Breakdown: Matrix Math
- **Reshaping**: Numpy's `.reshape()` changes the structure of data without changing the data itself. Using `-1` tells Numpy to 'figure out' that dimension automatically based on the length of the array.
- **Transposition**: `.T` flips the axes. In 1D arrays, this does nothing; the array must be at least 2D (like a column or row vector) for a visible change.
- **Dot Product vs Matrix Mult**: `np.dot` and `@` behave similarly for 2D matrices (performing standard matrix multiplication), but have different behaviors for higher-dimensional tensors.

### Explanation: Exercise 3 - Indexing and Masking

- **Multidimensional Indexing**: `mat[d, r, c]` accesses the depth, row, and column. Using `:` selects the entire range for that dimension.
- **Boolean Masking**: Expressions like `mat > 25` return a boolean array. When used as an index, NumPy returns a 1D array of only the `True` values.
- **Logical Operators**: Use `&` (AND) and `|` (OR) for element-wise boolean logic in NumPy. The standard Python `and`/`or` keywords will fail on arrays.
- **np.where(condition, x, y)**: This is a vectorized 'if-else'. It returns `x` where the condition is true and `y` where it is false.

```python
# 4. Performance vs Number of Neighbors
neighbor_range = range(1, 16)
train_accs = []
test_accs = []

for k in neighbor_range:
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train, y_train)
    train_accs.append(accuracy_score(y_train, clf.predict(X_train)))
    test_accs.append(accuracy_score(digits_test, clf.predict(images_test)))

plt.figure(figsize=(8, 4))
plt.plot(neighbor_range, train_accs, label='Train Accuracy', marker='o')
plt.plot(neighbor_range, test_accs, label='Test Accuracy', marker='s')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN Performance vs k')
plt.legend()
plt.grid(True)
plt.show()
```
